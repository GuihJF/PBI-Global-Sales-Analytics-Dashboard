```python
import pandas as pd

# Leitura do CSV
df = pd.read_csv('Sales_Dataset.csv', sep=';', encoding='utf-8')

# Renomear colunas
df.rename(columns={
    'ID_Pedido': 'order_id',
    'Data_Pedido': 'order_date',
    'ID_Cliente': 'client_id',
    'Segmento': 'segment',
    'Região': 'region',
    'País': 'country',
    'Categoria': 'category',
    'SubCategoria': 'subcategory',
    'Total_Vendas': 'total_sales',
    'Quantidade': 'quantity',
    'Desconto': 'discount',
    'Lucro': 'profit',
    'Prioridade': 'priority'
}, inplace=True)

# Traduzir valores categóricos
df['segment'] = df['segment'].replace({
    'Consumidor': 'Consumer',
    'Corporativo': 'Corporate',
    'Home Office': 'Home Office'
})

df['category'] = df['category'].replace({
    'Tecnologia': 'Technology',
    'Moveis': 'Furniture',
    'Suprimentos': 'Supplies'
})

df['subcategory'] = df['subcategory'].replace({
    'Accessories': 'Accessories',
    'Chairs': 'Chairs',
    'Phones': 'Phones',
    'Copiers': 'Copiers',
    'Tables': 'Tables',
    'Binders': 'Binders'
})

df['priority'] = df['priority'].replace({
    'Crítico': 'Critical',
    'Médio': 'Medium',
    'Alto': 'High',
    'Baixo': 'Low'
})

# Corrigir tipos
df['order_date'] = pd.to_datetime(df['order_date'], errors='coerce')
df['total_sales'] = pd.to_numeric(df['total_sales'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.'), errors='coerce')
df['profit'] = pd.to_numeric(df['profit'].str.replace('R$', '').str.replace('.', '').str.replace(',', '.'), errors='coerce')
df['discount'] = pd.to_numeric(df['discount'], errors='coerce')
df['quantity'] = pd.to_numeric(df['quantity'], errors='coerce')

# Exportar CSV traduzido
df.to_csv('Sales_Dataset_Translated.csv', index=False)
from google.colab import files
files.download('Sales_Dataset_Translated.csv')
```
